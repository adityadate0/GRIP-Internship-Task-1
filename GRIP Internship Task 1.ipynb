# Importing the important libraries

```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import sklearn.metrics
```

```python
student_data = pd.read_csv('https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv')
```

# Viewing the dataframe

```python
student_data.head(5)
```

# Splitting the data into training and testing set

```python
X = student_data['Hours']
Y = student_data['Scores']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.20,random_state=0)
```

# Checking the split data

```python
X_train
```

```python
Y_train
```

# Visualizing the data

```python
fig = plt.figure(figsize=(7,7));
ax = fig.add_subplot(111);
student_data.plot('Hours','Scores',style='o',ax = ax,color='red');
plt.title("Relation between Hours and Scores");
plt.ylabel("Scores");
```

# Reshaping data for model

```python
X_train = np.array(X_train).reshape(-1,1)
X_test = np.array(X_test).reshape(-1,1)
```

# Creating the linear regression model

```python
model = LinearRegression()
```

# Fitting the training data in the model

```python
model.fit(X_train,Y_train)
```

# Making predictions from test data

```python
predictions = model.predict(X_test)
```

# Evaluating the model

```python
model.score(X_test,Y_test)
```

# Comparing result with true values

```python
predictions1 = model.predict(np.array(X).reshape(-1,1))
```

```python
data_comparison = pd.DataFrame(data = zip(Y,predictions1),columns = ["True Score","Predicted Score"])
fig = plt.figure(figsize=(18,10))
axis = fig.add_subplot(111)
plot = data_comparison.plot(kind='line',ax = axis,)
plt.xticks(np.arange(25));
plt.legend(fontsize='x-large')
plt.xlabel("Student Number")
plt.ylabel('Score')
plt.title("Comparison of True and Predicted Values");
plt.style.use('seaborn-white')
```

# Getting performance scores

```python
def score_generator(X,Y):
    mae = sklearn.metrics.mean_absolute_error(X,Y)
    mse = sklearn.metrics.mean_squared_error(X,Y)
    r2_score = sklearn.metrics.r2_score(X,Y)
    print("Mean Absolute Error: {} | Mean Squared Error: {} | Accuracy Score: {}".format(mae,mse,r2_score))
```

```python
score_generator(Y_test,predictions)
```
